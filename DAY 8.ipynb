Basic import

```python
import math
result = math.sqrt(16)
print(result)
```

Import specific functions or Variables

```python
from math import sqrt,pi
print(sqrt(25))
print(pi)
```

Import with an Allas

```python
import math as m
print(m.sqrt(36))
```

Import All from a Module

```python
from math import *
print(sin(0))
```
