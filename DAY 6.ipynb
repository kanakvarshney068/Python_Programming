Variable-length positional arguments

```python
def sum_num(num1, *num):
  result = num1
  for i in num:
      result += i
  return result
r = sum_num(10,20,30)
print(f"The sum of numbers is: {r}")
r = sum_num(10,20,30,40,50)
print(f"The sum of numbers is: {r}")
```

Write a function to find the factorial of a number

```python
def factorial(n):
  if n == 0 or n == 1:
      return 1
  else:
    return n * factorial(n - 1)
num = 5
result = factorial(num)
print(f"Factorial of {num} is {result}")
```

write a function to check if a number is prime

```python
def is_prime(num):
  if num <= 1:
     return False
  for i in range(2, int(num**0.5) + 1):
   if num % i == 0:
     return False
  return True
print(is_prime(7))
```

Write a function to calculate the nth Fibonacci number using
recursion.

```python
def fibonacci(n):
  if n <= 1:
     return n
  return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(6))
```

Check Palindrome.

```python
def is_palindrome(s):
  return s == s[::-1]
# Test the function
print(is_palindrome("radar")) # Output: True
print(is_palindrome("hello"))
```

Write a function to Find GCD of Two Numbers

```python
def gcd(a, b):
  while b:
    a, b = b, a % b # a=b and b = a%b
  return a
# Test the function
print(gcd(12, 18))
```

Count Vowels in a String

```python
def count_vowels(s):
  vowels = "aeiouAEIOU"
  return sum(1 for char in s if char in vowels)
# Test the function
print(count_vowels("hello world"))
```

Check Armstrong Number

```python
def is_armstrong(n):
  num_str = str(n)
num_digits = len(num_str)
return sum(int(digit) ** num_digits for digit in num_str) == n
# Test the function
print(is_armstrong(153)) # Output: True
print(is_armstrong(123))
```
