Creating Sets

```python
empty_set =  set()
print(empty_set)
```

```python
#non empty set
my_set = {1,2,3}
print(my_set)
```

```python
#From a list
set_from_list = set([1,2,3])
print(set_from_list)
```

```python
#add(element)
my_set = {1,2,3,4,5}
my_set.add(6)
print(my_set)
```

```python
#remove(element)
my_set = {1,2,3,4,5}
my_set.remove(3)
print(my_set)
```

```python
my_set = {1,2,5,6}
my_set.remove(3)
print(my_set)
```

```python
# discard (element)
my_set = {1,2,5,6}
my_set.discard(3)
print(my_set)
```

```python
#pop()
my_set = {1,2,3,4,5,6}
removed = my_set.pop()
print(removed)
print(my_set)
```

```python
# clear()
my_set = {1,2,3,4,5,6}
my_set.clear()
print(my_set)
```

```python
#union (other_set)
set1 = {1,2,3,4,5}
set2 = {5,6,7,8,9}
union_set = set1.union(set2)
print(union_set)
```

```python
# intersection(other_set)
set1 = {1,2,3,4,5}
set2 = {5,6,7,8,9}
intersection_set = set1.intersection(set2)
print(intersection_set)
```

```python
#difference(other_set)
set1 = {1,2,3,4,5}
set2 = {5,6,7,8,9}
difference_set = set1.difference(set2)
print(difference_set)
```

```python
# symmetric_difference()
set1 = {1,2,3,4,5}
set2 = {5,6,7,8,9}
symmetric_difference_set = set1.symmetric_difference(set2)
print(symmetric_difference_set)
```

```python
#issubset(other_set)
set1 = {1,2,3}
set2 = {1,2,3,4,5}
is_subset = set1.issubset(set2)
print(is_subset)
```

```python
#issuperset(other_set)
set1 = {1,2,3}
set2 = {1,2,3,4,5}
is_superset = set2.issuperset(set1)
print(is_superset)
```

```python
#Example of set opertions
set1 = {1,2,3}
set2 = {3,4,5}
print(set1.union(set2))
print(set1.intersection(set2))
print(set1.difference(set2))
```

Dictionaries in python

```python
#Empty dictionary
empty_dict = {}
print(empty_dict)
```

```python
#Non - empty dictionary
my_dict = {"name": "John", "age": 30, "city": "New York"}
print(my_dict)
```

Using get(key,default)

```python
student = {"name": "John", "age": 25}
print(student.get("name"))  # Output: John
print(student.get("grade", "Not available"))
```

Using keys()

```python
keys = student.keys()
print(keys)
```

using values()

```python
values = student.values()
print(values)
```

Using items()

```python
items = student.items()
print(items)
```

Using update(other_dict)

```python
extra_info = {"grade": "A", "city": "New York"}
student.update(extra_info)
print(student)
```

Using pop(key)

```python
age = student.pop("age")
print(age)  # Output: 25
print(student)
```

Using popitem()

```python
last_item = student.popitem()
print(last_item)  # Output: ('city', 'New York')
print(student)
```

Using clear()

```python
student.clear()
print(student)
```

Loop over Sets

```python
my_set = {"apple", "banana", "Cherry"}
for item in my_set:
    print(item)
```

Loop over Dictionaries

```python
dict1 = {"a": 1, "b": 2, "c": 3}
for key in dict1:
  print(key)
```

```python
dict1 = {"a": 1, "b": 2, "c": 3}
for value in dict1.values():
  print(value)
```

```python
dict1 = {"a": 1, "b": 2, "c": 3}
for key, value in dict1.items():
  print(f"{key}: {value}")
```

Combined Example

```python
data = {"fruits": {"apple", "banana"}, "veggies":["carrot", "broccoli"]}
for category, items in data.items():
  print(f"{category}: {items}")
```
