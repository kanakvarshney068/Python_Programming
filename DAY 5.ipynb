check if a number is prime using a for loop and else

```python
num = int(input("Enter a number:"))
if num > 1:
  for i in range (2, int(num ** 0.5) + 1):
      if num % i == 0:
         print(f"{num} is not a prime number.")
         break
  else:
      print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

Print Fibonacci Series: 0,1,1,2,3,5,8,13.....

```python
n = int(input("Enter the number of terms for the fibonacci seies:"))
if n <= 0:
  print("Please enter a positive integer.")
else:
    a, b = 0,1
    print("Fibonacci Series:")
    for i in range(n):
      print(a, end=" ")
      a, b = b, a + b
```

Program to compute the factorial of a given number

```python
num = int(input("Enter a number to compute its factorial:"))
if num < 0:
  print("Factorial is not defined for negative numbers.")
elif num == 0:
  print("The factorial of 0 is 1.")
else:
  factorial = 1
  for i in range(1, num + 1):
      factorial *= i
  print(f"The factorial of {num} is {factorial}.")
```
